Now that we have some data to work on, lets turn the values into a format we can work with.

Among the four new lists, Skill level is not a numerical type of data. Lucky for us it seems to be ordinal, so we can replace it with numbers. Let us see what are the skill levels

```python
df['skill level'].unique()
```

Courses with Skill Level: General does not seem to be technical, I would likely to go with the following replacements:

General - 1

Beginner - 2

Beginner + Intermediate - 2.5

Intermediate - 3

Advanced - 4

```python
df['skill level num'] = df['skill level'].copy(deep = True)
df.head()
```

```python
for i in range(len(df['skill level num'])):
    level = df['skill level num'][i].split(": ")
    levelNo = 0.0
    if level[1] == 'General':
        levelNo = 1.0
    elif level[1] == 'Beginner':
        levelNo = 2.0
    elif level[1] == 'Beginner + Intermediate':
        levelNo = 2.5
    elif level[1] == 'Intermediate':
        levelNo = 3.0
    elif level[1] == 'Advanced':
        levelNo = 4.0
    df['skill level num'][i] = levelNo
```

```python
df["skill level num"] = df["skill level num"].astype(float)
```

There are three columns that we can turn into numerical data, namely the "viewer count", "release date" and "likes" columns. Let us work on those.

```python
df["viewer count"] = df["viewer count"].replace(" viewers", "", regex=True)
```

```python
df["viewer count"] = df["viewer count"].replace(",", "", regex=True)
```

```python
df["viewer count"] = df["viewer count"].astype(int)
```

```python
df.dtypes
```

```python
df.tail()
```

```python
df["likes"] = df["likes"].replace("Liked by", "", regex=True)
df["likes"] = df["likes"].replace(" users", "", regex=True)
df["likes"] = df["likes"].replace(",", "", regex=True)
```

```python
df['likes'] = df["likes"].replace(" user", "", regex=True)
```

```python
df['likes'] = df['likes'].fillna(0)
```

```python
df['likes'] = df['likes'].astype(int)
```

```python
df.tail()
```

```python
for i in range(len(df['duration'])):
    strings = df['duration'][i].split()
    minutes = 0
    for j in strings:
        if 'h' in j:
            minutes += int(j[:-1])*60
        if 'm' in j:
            minutes += int(j[:-1])
    df['duration'][i] = minutes
```

```python
df.tail()
```

```python
df["release date"] = df["release date"].replace("Released ", "", regex=True)
```

```python
df['release date'][0]
```

```python
df['release date'] = pd.to_datetime(
                          df['release date'],
                          format='%b %d, %Y'
)
```

```python
df['release date'][0]
```
